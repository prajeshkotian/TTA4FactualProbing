```python
!pip install googletrans==3.1.0a0

import googletrans
from googletrans import Translator
import json
import time

# Create an empty list to store prompt data
prompt_list = []

print(googletrans.LANGUAGES)

# Init


with open('wikidata12500.jsonl', 'r', encoding='utf-8') as infile:
    # Open a new file for writing
    translator = Translator()
    with open('chinese_prompt.jsonl', 'w', encoding='utf-8') as outfile:
        # Iterate over each line in the input file
        for line in infile:
            # Parse the JSON from the line
            data = json.loads(line)
            #time.sleep(1)
            output = translator.translate(data['os'], src='en', dest='zh-tw')
            output1 = translator.translate(data['ss'], src='en', dest='zh-tw')
            data['os'] = output.text
            data['ss'] = output1.text
            output2 = translator.translate(data['prompt'], src='en', dest='zh-tw')
            # Add "translated" to the prompt
            data['prompt'] = output2.text
            # Write the modified data to the output file
            outfile.write(json.dumps(data, ensure_ascii=False) + '\n')




# Print the list of prompt data
print(prompt_list)
```
